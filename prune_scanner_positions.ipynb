```python
# In this file, we prune locations that the readings for the areas in the s3dis dataset were taken
# The pruned locations are good candidates for the positions of the virtual scanners in Blensor
# The pruning is mostly about reducing positions too close to each other
# The locations are present in the locations.json file (for area 1)

%matplotlib notebook
import json
import numpy as np
import matplotlib.pyplot as plt

locations = []
with open("./locations.json") as f:
    locations = json.load(f)
    print("original len", len(locations))
    locations = list(set([tuple(x) for x in locations]))
    print("deduplicated len", len(locations))
    
xs = [l[0] for l in locations]
ys = [l[1] for l in locations]
zs = [l[2] for l in locations]


print(locations[:10])
plt.scatter(xs, ys)
plt.gca().set_aspect('equal')

plt.show()
    
```
