# Pruning scanner locations
In this file, we prune locations that the readings for the areas in the s3dis dataset were taken. The pruned locations are good candidates for the positions of the virtual scanners in Blensor. The pruning is mostly about reducing positions too close to each other. The original locations are present in the `locations/locationsX.json` file, where X is the number of the Area from the dataset. These were extracted using the `collect_camera_poses.py` script from the pose data in the s3dis dataset.

```python


%matplotlib notebook
import json
import numpy as np
import matplotlib.pyplot as plt


def prune(locations):
    pruned_locations = {}
    for k, v in locations.items():
        pruned_locations[k] = v[0]
    return pruned_locations

locations = []
with open("./locations6.json") as f:
    locations = json.load(f)
    
    
locations = prune(locations)
    
coords = locations.values()
xs = [l[0] for l in coords]
ys = [l[1] for l in coords]
zs = [l[2] for l in coords]

with open("./locations_pruned6.json", "w") as f:
    json.dump(locations, f)

plt.scatter(xs, ys)
plt.gca().set_aspect('equal')

plt.show()
    
```
