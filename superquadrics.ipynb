# Playing with superquadrics

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# scaling
A = 0.5
B = 1
C = 1

# parameters
r = 10
s = 10
t = 10

def f(omega, m):
    return np.sign(np.sin(omega))*np.abs(np.sin(omega))**m

def g(omega, m):
    return np.sign(np.cos(omega))*np.abs(np.cos(omega))**m

def x(u, v):
    return A*g(v, 2/r)*g(u, 2/r)
    
def y(u, v):
    return B*g(v, 2/s)*f(u, 2/s)

def z(u, v):
    return C*f(v,2/t)

u = np.linspace(0, 6, 100)
v = np.linspace(0, 6, 100)
U,V = np.meshgrid(u, v)

X = x(U, V)
Y = y(U, V)
Z = z(U, V)

# Display the mesh
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.set_xlim3d(-2, 2)
ax.set_ylim3d(-2, 2)
ax.set_zlim3d(-2, 2)
ax.plot_wireframe(X, Y, Z)
plt.show()
```
