# Visual comparison of error standard deviations between the real and virtual scans

```python
import numpy as np
import matplotlib.pyplot as plt
from thesis.utils import sorted_ids_by_angle

virtual_wall_stddevs = np.loadtxt("/home/branislav/repos/thesis/virtual_error_measurements/default_blensor_error_without_correction/wall.txt")
wall_stddevs = np.loadtxt("/home/branislav/repos/github_indoor++/pcds/error_measurements/deviations_by_laser_averaged/wall.txt")
```

```python
for dist in wall_stddevs:
    plt.plot(range(16), dist[sorted_ids_by_angle], "--")
plt.xticks(range(16))
plt.legend(range(5))

plt.show()

for dist in virtual_wall_stddevs:
    plt.plot(range(16), dist[sorted_ids_by_angle], "--")
plt.xticks(range(16))
plt.legend(range(5))
plt.show()

```

```python
for laser in wall_stddevs.T:
    plt.plot(range(1, 6), laser, "--")
plt.xticks(range(1, 6))
plt.show()

for laser in virtual_wall_stddevs.T:
    plt.plot(range(1, 6), laser, "--")
plt.xticks(range(1, 6))
plt.show()
```
